```python
    
class Queen(pieces):
    def __init__(self):
        pass
    
    #Queen can move all over the board as long as there are no obstacles
    def Moves(row,column):
        validMovesMatrix= [
            ["a","b","c","d","e","f","g","h"],
            [1,1,1,1,1,1,1,1], #row 1
            [1,1,1,1,1,1,1,1], #row 2
            [1,1,1,1,1,1,1,1], #row 3
            [1,1,1,1,1,1,1,1], #row 4
            [1,1,1,1,1,1,1,1], #row 5
            [1,1,1,1,1,1,1,1], #row 6
            [1,1,1,1,1,1,1,1], #row 7
            [1,1,1,1,1,1,1,1], #row 8
    ]
        
         #here i have to check whether this move is valid for this piece or not 
                        #and is it occupied or not
        if validMovesMatrix[row][column]==1:
            return 1
        
        return 0
    
    def validity(self,rowfrom, columnfrom, rowto, columnto):
        print("validity check:")
#         ##see if it moves within same column or row
        if rowfrom-1 == rowto and (columnfrom-1 == columnto or columnfrom == columnto or columnfrom+1 == columnto):
            return 1
        
        if rowfrom == rowto and (columnfrom-1 == columnto or columnfrom == columnto or columnfrom+1 == columnto): 
            return 1
        
        if rowfrom+1 == rowto and (columnfrom-1 == columnto or columnfrom == columnto or columnfrom+1 == columnto):
            return 1
        
        # checking the diagonals
        # we have to get the distance n that the queen will move to
        
        n = rowto - rowfrom
        m = columnto - columnfrom
        
#         all we need to check is whether |n| equals |m| or not
        if n == m:
            return 1
        if m == -n:
            return 1
        
        return 0

# trying our algorithm
rowfrom=1
columnfrom=1
rowto=3
columnto=4
queen =Queen()

if queen.validity(rowfrom,columnfrom,rowto,columnto):
    print("Valid")
else:
    print("Move is not Valid")

rowto=4

if queen.validity(rowfrom,columnfrom,rowto,columnto):
    print("Valid")
else:
    print("Move is not Valid")
```




As we can notice that:

The **Principle diagonal** is [Row  m, Column  m], where m is integer

and Row number is **directly proportional** to the Column number.

The **Secondary diagonal** is [Row + m, Column - m] (from up to down) and [Row - m, Column + m] (from down to up), where m is integer

and Row number is **inversely proportional** to the Column number.

    Let the Queen be at coordinates [R, C]

        [R  1, C  1]  	  	   [R-1, C+1] or [R+1, C-1]   	  	[R, C  1]   	  	[R  1, C]
        [R  2, C  2]           	[R-2, C+2] or [R+2, C-2]   	  	[R, C  2]  	  	 [R  2, C]
        [R  3, R  3]           	[R-3, C+3] or [R+3, C-3]   	  	[R, C  3]  	  	 [R  3, C]
        [R  n, C  n]           	[R-n, C+n] or [R+n, C-n]   	  	[R, C  n]  	  	 [R  n, C]
